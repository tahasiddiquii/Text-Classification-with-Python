```python
import pandas as pd
```

```python
df = pd.read_json("../data/4x60dataset.json")
df
```

```python
tmp = []

for col in df.columns:
    for row in df[col]:
        tmp.append((row, col))
        
df = pd.DataFrame(tmp, columns=["text", "label"])
df
```

```python
df.to_csv("../data/responses.csv")
```
